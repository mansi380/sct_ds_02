```python
import pandas as pd
import numpy as np

# Load Titanic dataset
 # Replace with your path to the Titanic dataset
df = pd.read_csv(r'C:\Users\admin\Desktop\New folder\Titaanic Datasetforlogisticregression .csv')

# Display the first few rows of the dataset
print(df.head())

# Check data types and missing values
print(df.info())

# Summary statistics of numerical columns
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Fill missing 'Age' values with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)


# Drop rows where the 'Survived' column is missing
df.dropna(subset=['Survived'], inplace=True)

# Check if any missing values remain
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of Fare
plt.figure(figsize=(8, 6))
sns.histplot(df['Fare'], kde=True, bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Count plot for Survival
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between Age and Survived
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Survival by Age')
plt.show()

# Relationship between Sex and Survived
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()

# Relationship between Pclass and Survived
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Pclass')
plt.show()

# Relationship between Fare and Survived
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Survival by Fare')
plt.show()
```

```python

correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Crosstab of Sex and Survived
sex_survival = pd.crosstab(df['Sex'], df['Survived'])
print(sex_survival)

# Crosstab of Pclass and Survived
pclass_survival = pd.crosstab(df['Pclass'], df['Survived'])
print(pclass_survival)
```
